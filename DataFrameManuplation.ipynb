```python
import pandas as pd
```

```python
df = pd.read_csv('coursera_resources/dataset/presidents.csv')
df.columns = [x.lower().strip() for x in df.columns]
df.head()
```

```python
df['first'] = df['president']
df['first'] = df['first'].replace('(\ \w*)$', "", regex=True)
df.head()
```

```python
del(df['first'])
df.head()
```

```python
def split_name(row):
    first = row['president'].split(" ")[0]
    last = row['president'].split(" ")[-1]
    
    row['first'] = first
    row['last'] = last
    return row

df = df.apply(split_name, axis='columns')
df.head()
```

```python

del(df['last'])
df.head()
```

```python
pattern = '(\w+)(?:\ )(\w+)'

df['president'].str.extract(pattern).head()
```

```python
pattern = '(?P<first>\w+)(?:\ )(?P<last>\w+)'

names = df['president'].str.extract(pattern)
names.head()
```

```python
df['first'] = names['first']
df['last'] = names['last']

df.head()
```

```python
pattern = '([\w]{3}\ [\w]{2},\ [\w]{4})'
df['born'] = df['born'].str.extract(pattern)
df.head()
```

```python
df['born'] = pd.to_datetime(df['born'])
df.head()
```
