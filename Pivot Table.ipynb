```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("resources/dataset/cwurData.csv")
df.head()
```

```python
def rank_level(group):
    if 1 <= group <= 100:
        return "First Tier Top University"
    elif 101 <= group <= 200:
        return "Second Tier Top University"
    elif 201 <= group <= 300:
        return "Third Tier Top University"
    else:
        return "Other Top University"

df["Rank Level"] = df["world_rank"].apply(lambda x: rank_level(x))
df.tail()
```

```python
df.pivot_table(values='score', index="country", columns = 'Rank Level', aggfunc=[np.mean]).head()
```

```python
df.pivot_table(values='score', index="country", columns = 'Rank Level', aggfunc=[np.mean, np.max]).head()
```

```python
df.pivot_table(values='score', index="country", columns = 'Rank Level', aggfunc=[np.mean, np.max], margins=True).head()
```
