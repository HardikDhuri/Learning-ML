```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('resources/dataset/census.csv')
```

```python
df = df[df['SUMLEV'] == 50]
df.head()
```

```python
%%timeit -n 3
for state in df['STNAME'].unique():
    avg = np.average(df.where(df['STNAME'] == state).dropna()['CENSUS2010POP'])
    print("Counties in state", state, "have an average population of", avg)
```

```python
%%timeit -n 3

for group, frame in df.groupby('STNAME'):
# groupby function returns a tuple where group is the key we are grouping by in our case state name
# and frame is the projected dataframe for that group
    avg = np.average(frame['CENSUS2010POP'])
    print("Counties in state", group, "have an average population of", avg)
```

```python
df = pd.read_csv("resources/dataset/listings.csv")
df.head()
```

```python
df = df.set_index(['cancellation_policy', 'review_scores_value'])
df.head()
```

```python
for group, frame in df.groupby(level=(0,1)):
    print(group)
```

```python
def separate_group(item):
    if item[1] == 10.0:
        return (item[0], "10.0")
    else:
        return (item[0], "not 10.0")

for group, frame in df.groupby(separate_group):
    print(group)
```

## Aggregation

```python
df.reset_index(inplace=True)
df.head()
```

```python
df.groupby("cancellation_policy").agg({"review_scores_value":np.average})
```

```python
df.groupby("cancellation_policy").agg({"review_scores_value" : (np.nanmean,np.nanstd),
                                      "reviews_per_month": (np.nanmean)})
```

```python
cols = ['cancellation_policy', 'review_scores_value']
transform_df = df[cols].groupby("cancellation_policy").transform(np.nanmean)
transform_df.head()
```

```python
transform_df.rename({'review_scores_value' : 'mean_review_scores'}, axis = 'columns', inplace=True)
```

```python
transform_df.head()
```

```python
df = df.merge(transform_df, left_index=True, right_index=True)
df.head()
```

## Filtering 

```python


df.groupby("cancellation_policy").filter(lambda x: np.nanmean(x['review_scores_value']) > 9.2).head()
```

```python
df = pd.read_csv("resources/dataset/listings.csv")
cols = ["cancellation_policy", "review_scores_value"]
df = df[cols]
df.head()
```

```python
def calc_mean_review_score(group):
    avg = np.nanmean(group['review_scores_value'])
    group['review_mean_scores'] = np.abs(avg - group['review_scores_value'])
    return group

df.groupby('cancellation_policy').apply(calc_mean_review_score).head()
```
